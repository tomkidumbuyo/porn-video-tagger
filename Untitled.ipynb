```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

BASE_URL = 'https://sxyprn.com'

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
max_count = 10
```

```python
def get_posts_from_page(page_link):
    response = requests.get(page_link)
    soup = BeautifulSoup(response.text, "html.parser")
    search_results = soup.find_all(class_="search_results")
    post_links = [BASE_URL +a["href"] for result in search_results for a in result.find_all("a", href=True) if a["href"].startswith("/post/")]
    print(post_links)
    return post_links
```

```python
def get_posts_from_category(category_link):
    post_links = []
    response = requests.get(category_link)
    soup = BeautifulSoup(response.text, "html.parser")
    searches_container = soup.find('div', {'id': 'center_control'})
    pagination_elements = searches_container.find_all('a')
    pagination_element_links = [BASE_URL + elem.get('href') for elem in pagination_elements]
    for pagination_element_link in pagination_element_links:
        post_links += get_posts_from_page(pagination_element_link)
    return post_links
```

```python
def get_post_links():
    post_links = []
    for i in range(0, 26701, 150):
        response = requests.get(BASE_URL + '/searches/' + str(i) + '.html')
        soup = BeautifulSoup(response.text, "html.parser")
        searches_container = soup.find('div', {'id': 'searches_container'})
        category_link_elements = searches_container.find_all('a')
        
        for category_link_element in category_link_elements:
            category_link = BASE_URL + category_link_element.get('href')
            post_links += get_posts_from_category(category_link)
    return post_links
            
```

```python
links = list(set(get_post_links()))
df = pd.DataFrame(links, columns=["links"])
df.to_csv("links.csv", index=False) 
```
